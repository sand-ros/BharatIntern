```python
import numpy as np
import pandas as pd
```

```python
movies=pd.read_csv('/content/TMBD Movie Dataset.csv')
```

```python
movies
```

```python
movies.head(5)
```

```python
movies.describe()
```

```python
movies.isnull().sum()
```

```python
movies['year'] = pd.DatetimeIndex(movies['release_date']).year
movies['month'] = pd.DatetimeIndex(movies['release_date']).month
movies
```

```python
movies.columns
```

```python
movies=movies[['id','overview','genres','popularity']]
```

```python
movies
```

```python
movies.loc[:, 'tags'] = movies['overview'] + movies['genres']
```

```python

new_data = movies.drop(columns=['overview', 'genres'])

print(new_data)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=10000, stop_words='english')
```

```python
cv
```

```python
vector=cv.fit_transform(new_data['tags'].values.astype('U')).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
similarity
```
